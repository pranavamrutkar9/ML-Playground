```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers
import pandas as pd
```

```python
np.random.seed(42)

data = {
    "monthly_salary_k": np.random.uniform(3, 12, 50),
    "years_at_company": np.random.randint(1, 11, 50),
    "overtime_hrs": np.random.uniform(0, 5, 50),
    "satisfaction": np.random.randint(1, 11, 50),
    "distance_km": np.random.randint(1, 45, 50)
}

df = pd.DataFrame(data)
logic = (
    ((df['satisfaction'] < 4) & (df['overtime_hrs'] > 2.5)) |
    ((df['distance_km'] > 30) & (df['monthly_salary_k'] < 5)) |
    (df['overtime_hrs'] > 4.5)
)
df['left_company'] = logic.astype(int)

noise_indices = np.random.choice(df.index, size=5, replace=False)
df.loc[noise_indices, 'left_company'] = 1 - df.loc[noise_indices, 'left_company']
```

```python
df
```

```python
X = df[['monthly_salary_k',	'years_at_company',	'overtime_hrs',	'satisfaction',	'distance_km'	]]
y = df['left_company']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)
```

```python
model = keras.Sequential([
    layers.Input(shape=(X_train.shape[1],)),
    layers.Dense(16, activation='relu'), # Increased neurons
    layers.Dense(8, activation='relu'),  # Added another hidden layer
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from sklearn.utils import validation
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=100, batch_size=10, verbose=1
)
```
