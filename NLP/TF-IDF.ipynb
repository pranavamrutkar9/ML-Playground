```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "I love Winter",
    "Winter is the best",
    "I love Spring",
    "Spring is great",
    "Spring and summer are great and best",
    "spring is great and summer is great",
    "rainy is great"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

print(f"Vocabulary: {vectorizer.get_feature_names_out()}\n")
print(f"Document-term matrix:\n{X.toarray()}")
```
