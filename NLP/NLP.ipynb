```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/NLP/train.txt', sep = ';', header=None, names = ['text', 'emotion'])
```

```python
unique_emotions = df['emotion'].unique()
emotion_numbers = {}
i = 0
for emo in unique_emotions:
  emotion_numbers[emo] = i
  i+=1

df['emotion'] = df['emotion'].map(emotion_numbers)
```

```python
df.head()
```

```python
df['text'] = df['text'].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
import string

def remove_punctuation(text):
  return text.translate(str.maketrans('', '', string.punctuation))
```

```python
df['text'] = df['text'].apply(remove_punctuation)
```

```python
df.head()
```

```python
def remove_numbers(text):
  new = ""
  for i in text:
    if not i.isdigit():
      new = new+i

  return new

df['text']=df['text'].apply(remove_numbers)
```

```python
df.head()
```

```python
import re

def remove_urls(text):
  url_pattern = re.compile(r'https?://\S+|www\.\S+')
  return url_pattern.sub(r'', text)

df['text'] = df['text'].apply(remove_urls)
```

```python
df.head()
```

```python
import re

def remove_html_tags(text):
  clean = re.compile('<.*?>')
  return re.sub(clean, '', text)

df['text'] = df['text'].apply(remove_html_tags)
```

```python
df.head()
```

```python
def remove_emojis(text):
  new = ""
  for i in text:
    if i.isascii():
      new = new+i

  return new

df['text'] = df['text'].apply(remove_emojis)
```

```python
df.head()
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df.loc[1]['text']
```

```python
word_tokenize(df.loc[1]['text'])
```

```python
def remove(txt):
  words = word_tokenize(txt)
  cleaned = []
  for i in words:
    if not i in stop_words:
      cleaned.append(i)

  return ' '.join(cleaned)
```

```python
df['text'] = df['text'].apply(remove)
df['text'].head()
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

bow_vectorizer = CountVectorizer()
tfidf_vectorizer = TfidfVectorizer()
```

```python
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

nb_model_bow = MultinomialNB()
nb_model_bow.fit(X_train_bow, y_train)

pred_nb_bow = nb_model_bow.predict(X_test_bow)

print("Accuracy:", accuracy_score(y_test, pred_nb_bow))
```

```python
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

nb_model_tfidf = MultinomialNB()
nb_model_tfidf.fit(X_train_tfidf, y_train)

pred_nb_tfidf = nb_model_tfidf.predict(X_test_tfidf)

print("Accuracy:", accuracy_score(y_test, pred_nb_tfidf))
```

```python
from sklearn.linear_model import LogisticRegression
lr_model_bow = LogisticRegression()
lr_model_bow.fit(X_train_bow, y_train)
pred_lr_bow = lr_model_bow.predict(X_test_bow)
print("Accuracy:", accuracy_score(y_test, pred_lr_bow))

lr_model_tfidf = LogisticRegression()
lr_model_tfidf.fit(X_train_tfidf, y_train)
pred_lr_tfidf = lr_model_tfidf.predict(X_test_tfidf)
print("Accuracy:", accuracy_score(y_test, pred_lr_tfidf))
```

```python
from sklearn.svm import SVC
svm_model_bow = SVC()
svm_model_bow.fit(X_train_bow, y_train)
pred_svm_bow = svm_model_bow.predict(X_test_bow)
print("Accuracy:", accuracy_score(y_test, pred_svm_bow))

svm_model_tfidf = SVC()
svm_model_tfidf.fit(X_train_tfidf, y_train)
pred_svm_tfidf = svm_model_tfidf.predict(X_test_tfidf)
print("Accuracy:", accuracy_score(y_test, pred_svm_tfidf))
```
