```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "I love Winter",
    "Winter is the best",
    "I love Spring",
    "Spring is great",
    "Spring and summer are great and best",
    "spring is great and summer is great"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

print(f"Vocabulary: {vectorizer.get_feature_names_out()}")
print(f"Document-term matrix:\n{X.toarray()}")
```
